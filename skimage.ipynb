```python
import skimage
from skimage import data, io
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
images = np.load('/content/drive/MyDrive/CS3244/deep_weeds/small_images_np.npy')
```

```python
exampleImage = images[0]
io.imshow(exampleImage)
plt.show()
```

```python
from skimage import color 
grayscale = color.rgb2gray(images[0]) 

io.imshow(grayscale)
plt.show()
```

```python
from skimage.filters import threshold_otsu 

exampleImage_gray = color.rgb2gray(exampleImage) 
threshold_value = threshold_otsu(exampleImage_gray)  

exampleImage_gray_threholded = exampleImage_gray > threshold_value  

# Show the original image
io.imshow(exampleImage_gray)
plt.show()

# Show the thresholded image
io.imshow(exampleImage_gray_threholded)
plt.show()
```

```python
from skimage.feature import canny 
exampleImage_gray = color.rgb2gray(exampleImage) 
 
# Apply Canny detector 
exampleImage_gray_edges = canny(exampleImage_gray) 

io.imshow(exampleImage_gray)
plt.show()

io.imshow(exampleImage_gray_edges)
plt.show()
```

```python
exampleImage_gray_threholded_edges = canny(exampleImage_gray_threholded) 

io.imshow(exampleImage_gray_threholded)
plt.show()

io.imshow(exampleImage_gray_threholded_edges)
plt.show()
```
